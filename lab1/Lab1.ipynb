```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Control policy Exercise 1

```python
def control(t, y):
    major = 4  
    minor = 2  
    ux = -major * np.sin(t)
    uy = minor * np.cos(t)
    return np.array([ux, uy])
```

#### Circular motion

```python
def control1(t, y):
    ux = np.cos(t)
    uy = np.sin(t)
    return np.array([ux, uy])
```

#### Control policy exercise 3

```python
def control3(t, y):
    sx = 1
    sy = 1.8
    ux = sx*(-np.sin(t))
    uy = sy*(np.cos(t)**2 - np.sin(t)**2)
    return np.array([ux, uy])
```

#### Integration

```python
def simulate(Δt, x, u, wind_std):
    wind_value = np.random.normal(0, wind_std, size=x.shape)
    x += Δt * (u + wind_value)
    return x
```

#### Sensor

```python
def sense(x):
    return x
```

#### Simulation

```python
tf = 10
Δt = 0.1
time = np.linspace(0, tf, int(tf / Δt) + 1)

x = np.array([0.,0.]) 
position = [np.copy(x)] 
# windc = 0.1
wind_std = 0.1
for t in time:
    y = sense(x) 
    u = control(t, y) # Control policy 1
    # u = control3(t, y) # Control policy 3
    x = simulate(Δt, x, u, wind_std) 
    position.append(np.copy(x))

position_array = np.array(position)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.grid()
plt.plot(position_array[:,0], position_array[:,1])
# print(position_array)
```
